```python
# Series
import pandas as pd
a = {"name1":"dhruv","name2":"ravi","name3":"jash" }
data = pd.Series(a)
print(data)
```

```python
x = [2,3,4,5,6,7]
var = pd.Series(x,index=['a','b','c','d','e','f'])
print(var)
print(var[0])
print(type(var))
```

```python
import pandas as pd

data = { 'name':['dhruv','prachi','ravi','himali'],'port':[11,12,3,2],'rank':[1,2,1,2]}
var = pd.DataFrame(data)
print(var,'\n')
print(var['name'])
```

```python
dic = { 'name':['soham','jash','ravi','dhruv'],'port':[11,12,3,2],'rank':[1,2,1,2]}
data2= pd.DataFrame(dic)  
print(data2 , '\n')
data2['total'] = data2['port'] + data2['rank']
print(data2)
```

```python
x = [2,3,4,5,6,7]
var = pd.Series(x,index=['a','b','c','d','e','f'],name="Python",dtype="float")
print(var)
print(var[0])
print(type(var))
```
