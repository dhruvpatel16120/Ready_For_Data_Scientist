Project 1: Basic calculator

Note: it is useful to understand the basic concept of python like ( input,output,if-elif-else statements,function,Arathmatic oprator,explict type casting,return statement,variables,try-catch error handling)

```python
def calculate(num1, num2,op):
    if op == "+":
        return num1 + num2
    
    elif op == "-":
        return num1 - num2
    
    elif op == "*":
        return num1 * num2
    
    elif op == "/":
        if num2 != 0:
            return num1 / num2
        else:
            return "Error: Division by zero"
    
    elif op == "%":
        return num1 % num2
    
    elif op == "**":
        return num1 ** num2
    
    else :
        return "Invalid operation"


try:
    num1 = int(input("enter number 1:"))
    num2 = int(input("enter number 2:"))
    op = input("enter operation (+, -, *, /,%,**): ")
    print("Output : ",calculate(num1, num2, op))

except ValueError:
    print("Invalid input. Please enter a valid integer.")

except:
    print("An unexpected error occurred.")
```

Project 2a:- find minimum and maximum from input data
note : learn logical oprator

```python
def find_min(n1,n2,n3):
    return (n1 if (n1<=n2 and n1<=n3) else (n2 if (n2<=n1 and n2<=n3) else n3))

def find_max(n1,n2,n3):
    return (n1 if (n1>=n2 and n1>=n3)else (n2 if (n2>=n1 and n2>=n3) else n3))

def main():
    try :
        n1 = int(input("Enter first number: "))
        n2 = int(input("Enter second number: "))
        n3 = int(input("Enter third number: "))
        print("enter your choice (min / max): ")
        choice = input().strip().lower()

        if choice == "min":
            print("Minimum is:", find_min(n1, n2, n3))
        elif choice == "max":
            print("Maximum is:", find_max(n1, n2, n3))
        else:
            print("Invalid choice. Please enter 'min' or 'max'.")

    except ValueError:
        print("Invalid input. Please enter a valid integer.") 
        
    except:
        print("An unexpected error occurred.")
main()
```
