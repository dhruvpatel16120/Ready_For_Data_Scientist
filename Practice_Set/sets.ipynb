```python
#sets

a = {1,2,3,4}
b = {2,4,6,8}
c = {1,3,5,7}

print("a:",a)
print("b:",b)
print("c:",c)
print("union assoication:")
print(a.union(b))
print(b.union(a))
print(" ")
print("intersection assoication:")
print(a.intersection(b))
print(b.intersection(a))
print(" ")
print("distributive Law:")
b_or_c = b.union(c)
print("B U C:",b_or_c)
print("A:",a)
print("A intersection (B U C) :",a.intersection(b_or_c))
a_and_b = a.intersection(b)
a_and_c = a.intersection(c)
print(" ")
print(" ")
print("a.intersection(b) :",a_and_b)
print("a.intersection(c):",a_and_c)
print("(a_and_b) U (a_and_c):",a_and_b.union(a_and_c))
```
